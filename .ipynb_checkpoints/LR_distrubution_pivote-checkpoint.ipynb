```julia
begin
	using LinearAlgebra
	m = 10
	A = rand(m,m)
	Aₒ = copy(A)
end;
```

```julia
function LRpivot!(A)
    m = size(A,1)                           # takes number rows 
    p = collect(1:m)                        # Initialisierung des Permutationsvektors
    for k = 1:m    
        j = argmax(abs.(A[k:m,k])) + (k-1)  # Index der Pivotzeile (we add (k-1) because we look at the elements below )
        p[[k,j]] = p[[j,k]]
        A[[k,j],:] = A[[j,k],:]             # Zeilenvertauschung
        A[k+1:m,k] /= A[k,k]                      
        A[k+1:m,k+1:m] -= A[k+1:m,k]*A[k:k,k+1:m] 
    end
    return p                  
end;

begin
	p = LRpivot!(A) 
	L = UnitLowerTriangular(A)  
	R = UpperTriangular(A)      
	@debug "" Aₒ[p,:] ≈ L*R
	@debug "" p
end
```
