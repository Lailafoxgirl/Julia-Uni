```julia
function my_cholesky(A)
    m = size(A,1)
    L = zeros(eltype(A),m,m)
    for k = 1:m
        lk = L[1:k-1,1:k-1]\A[1:k-1,k] 		# ℓₖ = (Lₖ₋₁)⁻¹aₖ
        L[k,1:k-1] = lk'
        L[k,k] = sqrt(A[k,k] - lk'*lk)		# λₖ = √(αₖ-ℓₖ'ℓₖ)
    end
    return L
end;

begin
	m = 100
	A = randn(m,m)
	A = (A+A')/2             # symmetrisiere
	A = A + (sqrt(2*m)+1)*I  # garantiert positiv definit 
	L = my_cholesky(A)
	@debug "" A ≈ L*L'
end

begin
	@time my_cholesky(A)
	@time F = cholesky(A)
	@debug "" L ≈ F.L
end
```
